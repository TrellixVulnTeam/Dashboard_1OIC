- プレーしたピッチのコーナーの座標

```python
def get_corners(match_data):
    """
    Returns a 4x2 matrix. (p1(x,y), p2(x,y),...)
    """
    yam = match_data.yaml
    
    mat = np.zeros((4,2))
    for i, key in enumerate(['p1', 'p2', 'p3', 'p4']):
        lat, lon = yam[key].split(',')
        mat[i,0], mat[i,1] = lon, lat
        
    return mat

# コーナーの位置情報を呼び出す関数
point_array = get_corners(match_data)
print(point_array)
```
